```python
df = df_null_removed.copy()
```

```python
df['MS SubClass'] = df['MS SubClass'].apply(str)
```

```python
df_obj_col = df.select_dtypes(include='object')
df_num_col = df.select_dtypes(exclude='object')
```

```python
df_num_col.info()
```

```python
df_obj_dummies = pd.get_dummies(df_obj_col,drop_first=True)
```

```python
df = pd.concat([df_num_col,df_obj_dummies],axis=1)
```

```python
df.isnull().sum().sort_values()
```

```python
X = df.drop('SalePrice',axis=1)
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_model = ElasticNet()
```

```python
param_grid = {'alpha':[0.1,1,5,10,100],
              'l1_ratio':[0.1,0.7,0.99,1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(base_model,param_grid=param_grid,scoring='neg_mean_squared_error',cv=5)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_params_
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
def report(model):
    preds = model.predict(X_test)
    print(f'MAE: {mean_absolute_error(y_test,preds)}')
    print(f'RMSE: {np.sqrt(mean_squared_error(y_test,preds))}')
```

```python
report(grid_model)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
rf_model = RandomForestRegressor()
```

```python
report(rf_grid_model)
```

```python
tf_X = X.values
tf_y = y.values
```

```python
tf_X_train, tf_X_test, tf_y_train, tf_y_test = train_test_split(tf_X, tf_y, test_size=0.1, random_state=101)
```

```python
scaled_tf_X_train = scaler.fit_transform(tf_X_train)
scaled_tf_X_test = scaler.transform(tf_X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
scaled_tf_X_train.shape
```

```python
model = Sequential()
model.add(Dense(units=273,activation='relu'))
model.add(Dense(units=256,activation='relu'))
model.add(Dense(units=128,activation='relu'))
model.add(Dense(units=64,activation='relu'))
model.add(Dense(units=1,activation='relu'))

model.compile(optimizer='adam',loss='mse')
```

```python
early_stop = EarlyStopping(patience=5)
```

```python
model.fit(scaled_tf_X_train,tf_y_train,validation_data=(scaled_tf_X_test,tf_y_test),epochs=20000,callbacks=[early_stop])
```

```python
loss = pd.DataFrame(model.history.history)
loss.plot()
```

```python
report(model)
```
