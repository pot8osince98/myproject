```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(class_weight='balanced')
```

```python
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay,make_scorer,recall_score
```

```python
print(classification_report(y_test,preds))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,preds)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
scorer = make_scorer(recall_score,pos_label=0)
```

```python
k_model = KNeighborsClassifier()
```

```python
k_model.fit(X_train,y_train)
```

```python
pred = k_model.predict(X_test)
```

```python
print(classification_report(y_test,pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,pred)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
log_model.fit(X_train,y_train)
```

```python
log_preds = log_model.predict(X_test)
```

```python
print(classification_report(y_test,log_preds))
```

```python
k_values = list(range(1,20))
param_grid = {'n_neighbors': k_values}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(k_model,param_grid,scoring=scorer)
```

```python
grid.fit(X_train,y_train)
```

```python
k_preds = grid.predict(X_test)
```

```python
print(classification_report(y_test,k_preds))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,k_preds)
```

```python
grid = GridSearchCV(k_model,param_grid,scoring='recall')
```

```python
grid.fit(X_train,y_train)
```

```python
k_preds_two = grid.predict(X_test)
```

```python
print(classification_report(y_test,k_preds_two))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,k_preds_two)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(patience=5,monitor='val_recall_4')
```

```python
model.fit(x=X_train,y=y_train,epochs=300000,callbacks=[early_stop],validation_data=(X_test,y_test))
```

```python
model2 = Sequential()
model2.add(Dense(units=29,activation='relu'))
model2.add(Dense(units=128,activation='relu'))
model2.add(Dense(units=1,activation='sigmoid'))
model2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy',Recall()])
```

```python
model2.fit(x=X_train,y=y_train,epochs=300000,callbacks=[early_stop],validation_data=(X_test,y_test))
```

```python
plt.style.use('default')
plt.figure(figsize=(12,8))
preds = (model2.predict(X_test))
preds = np.where(preds>0.1,1,0)
ConfusionMatrixDisplay.from_predictions(y_test,preds)
plt.savefig('../static/credit_card/confusionmatrix.png')
```

```python
model3 = Sequential()
model3.add(Dense(units=29,activation='relu'))
model3.add(Dense(units=64,activation='relu'))
model3.add(Dense(units=32,activation='relu'))
model3.add(Dense(units=1,activation='sigmoid'))
model3.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy',Recall()])
```

```python
model3.fit(x=X_train,y=y_train,epochs=300000,callbacks=[early_stop],validation_data=(X_test,y_test))
```

```python
plt.style.use('default')
plt.figure(figsize=(12,8))
preds = (model3.predict(X_test))
preds = np.where(preds>0.1,1,0)
ConfusionMatrixDisplay.from_predictions(y_test,preds)
plt.savefig('../static/credit_card/confusionmatrix.png')
```
